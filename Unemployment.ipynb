```python
import pandas as pd
import numpy as np

unemployment_data=pd.read_csv("/content/Unemployment in India.csv")
print(unemployment_data)
```

```python

df = pd.DataFrame(unemployment_data)

df[" Estimated Unemployment Rate (%)"].replace(np.nan, 0, inplace=True)
df[" Date"].replace(np.nan, 0, inplace=True)
df[" Estimated Employed"].replace(np.nan, 0, inplace=True)
df[" Estimated Labour Participation Rate (%)"].replace(np.nan, 0, inplace=True)
df["Area"].replace(np.nan, 0, inplace=True)
df[" Frequency"].replace(np.nan, 0, inplace=True)
df["Region"].replace(np.nan, 0, inplace=True)
print(df)
```

```python
import matplotlib.pyplot as plt
unemployment_data.plot(kind="line",x=" Date",y=" Estimated Unemployment Rate (%)")
plt.title("unemployment rate")
plt.xlabel("date")
plt.ylabel("unemployment rate")
plt.show()
```

```python
unemployment_data["covid_period"] = np.where(unemployment_data[" Date"] < "2020-03-01", "pre_covid", "post_covid")
```

```python
post_covid_unemployment = unemployment_data[unemployment_data["covid_period"] == "post_covid"][" Estimated Unemployment Rate (%)"]

print("Post-Covid Unemployment Rate:", post_covid_unemployment.mean())
```
